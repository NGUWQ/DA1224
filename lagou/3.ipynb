```python
#箱型图(经验与工资的关系)
import os
import pandas as pd
from pyecharts import Bar,Boxplot
os.chdir('C:/爬虫/拉勾网')

data=pd.read_excel('C:/爬虫/拉勾网/数据分析岗位finals.xlsx')
grouped=data.groupby(['经验'])
grouped_pct=grouped['aSalary']
aprice=grouped_pct.agg(['mean','count'])

y1=data['aSalary'][data['经验']=='经验1-3年 ']
y2=data['aSalary'][data['经验']=='经验10年以上 ']
y3=data['aSalary'][data['经验']=='经验1年以下 ']
y4=data['aSalary'][data['经验']=='经验3-5年 ']
y5=data['aSalary'][data['经验']=='经验5-10年 ']
y6=data['aSalary'][data['经验']=='经验不限 ']
y7=data['aSalary'][data['经验']=='经验应届毕业生 ']

boxplot = Boxplot("工资与经验关系透视")
x_axis = ['经验1-3年', '经验10年以上', '经验1年以下', '经验3-5年', '经验5-10年', '经验不限', '经验应届毕业生']
y_axis = [
    y1,
    y2,
    y3,
    y4,
    y5,
    y6,
    y7,
]
_yaxis = boxplot.prepare_data(y_axis)       # 转换数据
boxplot.add("月薪(元)", x_axis, _yaxis,xaxis_rotate=30)
boxplot.render()
boxplot
aprice
attr = ['经验1-3年', '经验10年以上', '经验1年以下', '经验3-5年', '经验5-10年', '经验不限', '经验应届毕业生']
aprice['mean']=round(aprice['mean'],0)
v1 = aprice['mean']
bar = Bar("工资与经验关系")
bar.use_theme("infographic")
bar.add("月平均工资(元)", attr, v1, is_stack=True,is_label_show=True,xaxis_rotate=30)
bar.render()
bar
```

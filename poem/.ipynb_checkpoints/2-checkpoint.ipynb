```python
#-*- coding:utf-8 -*-
import pandas as pd
from pyecharts import Scatter

data=pd.read_excel('c:/daima/dataanalysis/poem/100.xlsx')
data['dynasty'][0:2]#近代
data['dynasty'][2:8]#两汉
data['dynasty'][8:15]#明代
data['dynasty'][15:19]#南北朝
data['dynasty'][19:22]#清代
data['dynasty'][22:51]#宋代
data['dynasty'][51:85]#唐代
data['dynasty'][85:87]#魏晋
data['dynasty'][87:89]#五代
data['dynasty'][89:96]#先秦
data['dynasty'][96:100]#元代



v1=data['count'][0:2]
v2=data['good'][0:2]
v3=data['count'][2:8]
v4=data['good'][2:8]
v5=data['count'][8:15]
v6=data['good'][8:15]
v7=data['count'][15:19]
v8=data['good'][15:19]
v9=data['count'][19:23]
v10=data['good'][19:23]
v11=data['count'][23:51]
v12=data['good'][23:51]
v13=data['count'][51:86]
v14=data['good'][51:86]
v15=data['count'][86:88]
v16=data['good'][86:88]
v17=data['count'][88:90]
v18=data['good'][88:90]
v19=data['count'][90:96]
v20=data['good'][90:96]
v21=data['count'][96:100]
v22=data['good'][96:100]
scatter = Scatter()
scatter.add('近代', v1, v2)
scatter.add('两汉', v3, v4)
scatter.add('明代', v5, v6)
scatter.add('南北朝', v7, v8)
scatter.add('清代', v9, v10)
scatter.add('宋代', v11, v12)
scatter.add('唐代', v13, v14)
scatter.add('魏晋', v15, v16)
scatter.add('五代', v17, v18)
scatter.add('先秦', v19, v20)
scatter.add('元代', v21, v22,yaxis_min=8)
scatter.render()
scatter


```
